```python
import pandas as pd
import numpy as np

ser=pd.Series(np.take(list('abcdefgh'),np.random.randint(8,size=30)))
ser
```

list('abcdefgh') → tạo danh sách ký tự ['a','b','c','d','e','f','g','h']

np.random.randint(8, size=30) → sinh 30 số ngẫu nhiên từ 0 đến 7

np.take(..., ...) → lấy các ký tự trong danh sách theo các chỉ số ngẫu nhiên đó

pd.Series(...) → chuyển kết quả thành Pandas Series

```python
so_lan=ser.value_counts().reset_index(name='so lan')
print(so_lan)
```
