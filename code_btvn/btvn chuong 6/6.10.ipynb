```python
import threading
import random

lst = [random.randint(0, 100) for _ in range(20)]
max_values = []

def find_max(part):
    max_values.append(max(part))

mid = len(lst) // 2

t1 = threading.Thread(target=find_max, args=(lst[:mid],))
t2 = threading.Thread(target=find_max, args=(lst[mid:],))

t1.start()
t2.start()

t1.join()
t2.join()

print("List:", lst)
print("Max:", max(max_values))
```
