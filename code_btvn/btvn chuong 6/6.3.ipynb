```python
import threading

def even():
    for i in range(30, 51):
        if i % 2 == 0:
            print("Even:", i)

def odd():
    for i in range(30, 51):
        if i % 2 != 0:
            print("Odd:", i)

t1 = threading.Thread(target=even)
t2 = threading.Thread(target=odd)

t1.start()
t2.start()

t1.join()
t2.join()
```
