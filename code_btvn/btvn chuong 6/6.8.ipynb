```python
import threading
import random

lst = [random.randint(0, 10) for _ in range(20)]
total = 0

def sum_part(part):
    global total
    total += sum(part)

mid = len(lst) // 2

t1 = threading.Thread(target=sum_part, args=(lst[:mid],))
t2 = threading.Thread(target=sum_part, args=(lst[mid:],))

t1.start()
t2.start()

t1.join()
t2.join()

print("List:", lst)
print("Total:", total)
```
