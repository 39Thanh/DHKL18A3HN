```python
import threading
import time
import datetime

def print_time(name, delay, times):
    print("Starting", name)
    for _ in range(times):
        time.sleep(delay)
        print(name + ":", datetime.datetime.now())
    print("Exiting", name)

print("Starting Main Thread")

t1 = threading.Thread(target=print_time, args=("Google", 1, 3))
t2 = threading.Thread(target=print_time, args=("Yahoo", 2, 3))
t3 = threading.Thread(target=print_time, args=("Facebook", 3, 3))

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```
