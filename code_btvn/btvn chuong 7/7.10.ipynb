```python
import tkinter as tk
from tkinter import messagebox

def kiem_tra():
    ten = entry_name.get()
    tuoi = int(entry_age.get())
    if tuoi >= 18:
        messagebox.showinfo("Info", f"Xin chào {ten}\nBạn đã trưởng thành!")
    else:
        messagebox.showinfo("Info", f"Xin chào {ten}\nBạn còn nhỏ tuổi!")

root = tk.Tk()
root.title("Age Check")

tk.Label(root, text="Name:").pack()
entry_name = tk.Entry(root)
entry_name.pack()

tk.Label(root, text="Age:").pack()
entry_age = tk.Entry(root)
entry_age.pack()

tk.Button(root, text="Check", command=kiem_tra).pack()

root.mainloop()
```
