```python
import tkinter as tk

def tinh_tien():
    km = float(entry.get())
    if km <= 0.8:
        tien = 12000
    else:
        tien = 12000 + (km - 0.8) * 13000
    result.config(text=f"Taxi fee: {int(tien)} VND")

root = tk.Tk()
root.title("Taxi Fee")

tk.Label(root, text="Enter distance (km):").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Calculate", command=tinh_tien).pack()
result = tk.Label(root, text="")
result.pack()

root.mainloop()
```
