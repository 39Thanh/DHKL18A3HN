```python
import tkinter as tk
from tkinter import messagebox

def chon(i):
    messagebox.showinfo("Info", f"Bạn chọn bài {i}")

root = tk.Tk()
root.title("Menu")

menu = tk.Menu(root)
root.config(menu=menu)

chuong = tk.Menu(menu, tearoff=0)
menu.add_cascade(label="Chương 1", menu=chuong)

for i in range(1, 8):
    chuong.add_command(label=f"1.{i}", command=lambda i=i: chon(i))

root.mainloop()
```
