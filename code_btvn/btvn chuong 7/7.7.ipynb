```python
import tkinter as tk

def tinh_tong():
    n = int(entry.get())
    s = sum(range(1, n + 1))
    result.config(text=f"The sum is 1 + 2 + ... + {n} = {s}")

root = tk.Tk()
root.title("Sum of N")

tk.Label(root, text="Enter value of integer N:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Validate", command=tinh_tong).pack()
result = tk.Label(root, text="")
result.pack()

root.mainloop()
```
