```python
import tkinter as tk

def submit():
    print("Tên:", name.get())
    print("ID:", sid.get())
    print("Mật khẩu:", pwd.get())

root = tk.Tk()
root.title("Student Form")

tk.Label(root, text="Student Name").grid(row=0, column=0)
tk.Label(root, text="Student ID").grid(row=1, column=0)
tk.Label(root, text="Password").grid(row=2, column=0)

name = tk.Entry(root)
sid = tk.Entry(root)
pwd = tk.Entry(root, show="*")

name.grid(row=0, column=1)
sid.grid(row=1, column=1)
pwd.grid(row=2, column=1)

tk.Button(root, text="Submit", command=submit).grid(row=3, column=1)

root.mainloop()
```
