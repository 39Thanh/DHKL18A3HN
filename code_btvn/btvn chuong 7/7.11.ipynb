```python
import tkinter as tk
from tkinter import messagebox

can = ["Canh","Tân","Nhâm","Quý","Giáp","Ất","Bính","Đinh","Mậu","Kỷ"]
chi = ["Thân","Dậu","Tuất","Hợi","Tý","Sửu","Dần","Mão","Thìn","Tỵ","Ngọ","Mùi"]

def doi_nam():
    year = int(entry.get())
    name = can[year % 10] + " " + chi[year % 12]
    messagebox.showinfo("Result", f"Năm âm lịch: {name}")

root = tk.Tk()
root.title("Lunar Year")

tk.Label(root, text="Enter solar year:").pack()
entry = tk.Entry(root)
entry.pack()

tk.Button(root, text="Convert", command=doi_nam).pack()

root.mainloop()
```
