```python
import tkinter as tk
import math

def tinh():
    m = int(e1.get())
    n = int(e2.get())
    if var.get() == "GCD":
        kq = math.gcd(m, n)
        result.config(text=f"GCD(m, n) = {kq}")
    else:
        kq = m * n // math.gcd(m, n)
        result.config(text=f"LCM(m, n) = {kq}")

root = tk.Tk()
root.title("GCD - LCM")

tk.Label(root, text="Enter value of m:").pack()
e1 = tk.Entry(root)
e1.pack()

tk.Label(root, text="Enter value of n:").pack()
e2 = tk.Entry(root)
e2.pack()

var = tk.StringVar(value="GCD")
tk.OptionMenu(root, var, "GCD", "LCM").pack()

tk.Button(root, text="Validate", command=tinh).pack()
result = tk.Label(root, text="")
result.pack()

root.mainloop()
```
