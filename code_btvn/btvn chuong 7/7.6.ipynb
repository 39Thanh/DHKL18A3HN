```python
import tkinter as tk

def reverse():
    s = entry.get()
    r = ""
    for i in s:
        r = i + r
    result.config(text=r)

root = tk.Tk()
root.title("Reverse String")

tk.Label(root, text="Enter a word:").pack()
entry = tk.Entry(root, width=30)
entry.pack()

tk.Button(root, text="Validate", command=reverse).pack(pady=5)
result = tk.Label(root, text="")
result.pack()

root.mainloop()
```
