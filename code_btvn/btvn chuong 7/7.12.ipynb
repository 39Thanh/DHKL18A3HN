```python
import tkinter as tk
from tkinter import messagebox

def tinh_bmi():
    try:
        can_nang = float(entry_weight.get())
        chieu_cao = float(entry_height.get())

        bmi = can_nang / (chieu_cao ** 2)
        bmi = round(bmi, 2)

        if bmi < 18.5:
            ket_luan = "Gầy"
        elif bmi < 25:
            ket_luan = "Bình thường"
        elif bmi < 30:
            ket_luan = "Thừa cân"
        else:
            ket_luan = "Béo phì"

        label_bmi.config(text=f"BMI: {bmi}")
        label_ket_luan.config(text=f"Kết luận: {ket_luan}")

    except:
        messagebox.showerror("Lỗi", "Vui lòng nhập số hợp lệ")

# ====== Giao diện ======
root = tk.Tk()
root.title("BMI Calculator")
root.geometry("300x250")

tk.Label(root, text="Cân nặng (kg):").pack(pady=5)
entry_weight = tk.Entry(root)
entry_weight.pack()

tk.Label(root, text="Chiều cao (m):").pack(pady=5)
entry_height = tk.Entry(root)
entry_height.pack()

tk.Button(root, text="Tính BMI", command=tinh_bmi).pack(pady=10)

label_bmi = tk.Label(root, text="BMI:")
label_bmi.pack(pady=5)

label_ket_luan = tk.Label(root, text="Kết luận:")
label_ket_luan.pack(pady=5)

root.mainloop()
```
