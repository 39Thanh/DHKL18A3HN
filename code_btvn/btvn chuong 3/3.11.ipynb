```python
stocks = np.concatenate((stock1, stock2))
print(stocks[-15:])
```

#4

```python
stocks_companies = np.zeros(
    stocks.shape[0],
    dtype=stocks.dtype.descr + companies.dtype.descr
)

for i, row in enumerate(stocks):
    stocks_companies[i][stocks.dtype.names] = row
    company = companies[companies['name'] == row['symbol']]
    if len(company) > 0:
        stocks_companies[i][companies.dtype.names] = company[0]
stocks_companies[:5]
```

#5

```python
symbols = np.unique(stocks['symbol'])

result_avg = []

for sym in symbols:
    mask = stocks['symbol'] == sym
    result_avg.append((
        sym,
        np.mean(stocks['open'][mask]),
        np.mean(stocks['high'][mask]),
        np.mean(stocks['low'][mask]),
        np.mean(stocks['close'][mask]),
        np.mean(stocks['volume'][mask])
    ))

result_avg = np.array(
    result_avg,
    dtype=[
        ('symbol','U10'),
        ('open_mean',float),
        ('high_mean',float),
        ('low_mean',float),
        ('close_mean',float),
        ('volume_mean',float)
    ]
)

result_avg
```

#6

```python
result_close = []

for sym in symbols:
    mask = stocks['symbol'] == sym
    closes = stocks['close'][mask]
    result_close.append((
        sym,
        np.mean(closes),
        np.max(closes),
        np.min(closes)
    ))

result_close = np.array(
    result_close,
    dtype=[
        ('symbol','U10'),
        ('close_mean',float),
        ('close_max',float),
        ('close_min',float)
    ]
)

result_close
```
