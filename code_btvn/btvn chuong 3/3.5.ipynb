```python
import numpy as np
height=np.genfromtxt('heights_1.txt', delimiter=',', encoding='utf-8',skip_header=1,dtype=int)
weight=np.genfromtxt('weights_1.txt', delimiter=',',encoding='utf-8',skip_header=1,dtype=int)
```

```python
#1
arr_height=np.array(height)
print(arr_height)
```

```python
#2
arr_weight=np.array(weight)
print(arr_weight)
```

```python
#3
arr_height_m=arr_height*0.0254
print("du lieu sau khi doi tu inch sang m: \n", arr_height_m)
```

```python
#4
arr_weight_kg=arr_weight*0.453592
print("du lieu sau khi doi tu pound sang kg: \n", arr_weight_kg)
```

```python
#5
arr_bmi=arr_weight/(arr_height*arr_height)
print("BMI cua cac cau thu",arr_bmi)
```

```python
#6
can_nang_50=arr_weight_kg[49]
print("gia tri can nang o vi tri 50: ",can_nang_50)
```

```python
#7
arr_height_m_100=arr_height_m[100:109]
print(arr_height_m_100)
```

```python
#8
bmi_nho_hon_21=arr_bmi<21
print(bmi_nho_hon_21)
```

```python
#9
chieu_cao_tb=arr_height.mean(axis=0)
can_nang_tb=arr_weight.mean(axis=0)

print("chieu cao trung binh: ",chieu_cao_tb)
print("can nang trung binh: ",can_nang_tb)
```
