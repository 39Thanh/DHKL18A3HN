```python
import numpy as np 
 
# Định nghĩa kiểu dữ liệu cho một Structured Array 
dtype = [('name', 'U20'), ('age', int), ('height', float)] 
 
# Tạo một Structured Array 
data = np.array([('Minh', 22, 165.5), 
                 ('Lan', 40, 160.0), 
                 ('Linh', 31, 165.0), 
                 ('Lương',18,166.5), 
                 ('Thanh',50, 170.0)], dtype=dtype) 
 
# Tạo DataFrame từ Structured Array 
df = pd.DataFrame(data) 
 
# In ra DataFrame 
print(df) 
print(data.dtype)
```

```python
import pandas as pd 
indA=pd.Index([0, 2, 4, 6, 8]) 
indB=pd.Index([1, 3, 7, 8, 9]) 
 
print('Index A: ', indA) 
print('Index B: ', indB) 
print('Hợp: ', indA.union(indB))
print('Hiệu: ', indA.difference(indB)) 
print('Giao: ', indA.intersection(indB)) 
print('Hiệu đối xứng: ',indA.symetric_difference(indB))
print('A <= B and A!=B: ', indA.isin(indB))
```
